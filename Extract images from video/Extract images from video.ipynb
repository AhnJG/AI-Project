```python
# 영상 1초당 25~26장의 이미지 생성 (카카오 다운로드)
# Importing all necessary libraries 
import cv2 
import os 
```

```python
data_name = 'test_video_2.mp4'
data_path = os.getcwd()+'/data/'+data_name

#print(os.getcwd()+'/data/'+data_name)
#print(data_name[0:-4])
```

```python
dir_name = data_name[0:-4]

try:       
    # creating a folder named data 
    if not os.path.exists('./img/' + dir_name):
        os.makedirs('./img/' + dir_name) 
  
# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of ' + dir_name)
```

```python
# Read the video from specified path 
cam = cv2.VideoCapture(data_path) 
  
# frame 
currentframe = 0
  
while(True): 
      
    # reading from frame 
    ret,frame = cam.read() 
  
    if ret: 
        # if video is still left continue creating images 
        name = './img/'+ data_name[0:-4] + '/img'+ str(currentframe) + '.png'
        print ('Creating...' + name) 
  
        # writing the extracted images 
        cv2.imwrite(name, frame) 
  
        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break
  
# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows()
```
