```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

### haar cascade 
opencv 에서 제공하는 학습이 되어있는 검출기 ... 얼굴, 눈, 상반신, 하반신 등등

https://github.com/opencv/opencv/tree/master/data/haarcascades 에서 다운로드 (이제 opencv 에서 제공해서 다운 안해도 됨)

```python
# face_cascade = cv2.CascadeClassifier('./haar_cascade/haarcascade_frontalface_default.xml')
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

#### gray image

```python
image = cv2.imread('./image/bts1.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,8))
plt.imshow(gray_image, cmap='gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

#### face detection

```python
faces = face_cascade.detectMultiScale(gray_image, 1.3, 5)

print(faces.shape)
print("Number of faces detected: " + str(faces.shape[0]))
```

#### draw box

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)

plt.figure(figsize=(12,12))
plt.imshow(image, cmap='gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
